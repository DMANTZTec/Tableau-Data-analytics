```python
import pandas as pd
from math import pi
dt = pd.read_excel("AmulyaTechwork.xlsx", parse_dates=["STARTDATE"])
from bokeh.plotting import figure, show, output_file
from bokeh.embed import components
from bokeh.transform import cumsum
from bokeh.resources import CDN
from bokeh.models import ColumnDataSource
from bokeh.palettes import Category20c
from bokeh.palettes import Spectral6
```

```python
df = dt.drop("Comments",1)
df["MONTH"] = df["STARTDATE"].dt.strftime('%m')
df1= df[(df["MONTH"] == ("10"))]
df1
```

```python
e = df1.groupby(["SUBJECT"])["ACTUAL HOURS"].count()
e1 = e.index.tolist()
e1
```

```python
source = ColumnDataSource(data=dict(e1=e1, e=e, color=Spectral6))
```

```python
w = figure(x_range=e1, width=430, height=340, toolbar_location=None,
            tools="hover", tooltips="@e1: @e", title="noOfTasks of Amulya in 10th month")
```

```python
w.vbar(x="e1", top="e", width=0.6, color="color",legend_field="e1", source=source)

show(w)
```

```python
script3, div3 = components(w)
cdn_js = CDN.js_files[0]
cdn_js
```

```python
f = df.groupby(["SUBJECT", "NAME"])["ACTUAL HOURS"].sum()
f1 = pd.Series(f).reset_index(name="values")
f1["angle"] = f1["values"] / f1["values"].sum() *2*pi
f1["color"] = Category20c[len(f1)]
f1
```

```python
q = figure(height =350,  title="Pie Chart", toolbar_location=None,
           tools="hover", tooltips="@NAME,@SUBJECT: @values hrs", x_range=f1["SUBJECT"])
```

```python
q.wedge(x=2, y=0, radius=1.5,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color="color", legend_field="SUBJECT", source=f1)

q.axis.axis_label = None
q.axis.visible = False
q.grid.grid_line_color = None

show(q)
```
