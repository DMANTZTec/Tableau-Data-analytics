```python
import pandas as pd
from math import pi
r = pd.read_excel("mamathapriya.xlsx", parse_dates=["START DATE"])
from bokeh.plotting import figure, show, output_file
from bokeh.embed import components
from bokeh.transform import cumsum
from bokeh.resources import CDN
from bokeh.models import ColumnDataSource
from bokeh.palettes import Category20c
from bokeh.palettes import Spectral6
```

```python
r
```

```python
r1 = r.groupby(["SUBJECT", "lm"])["MODULE"].count()
r2 = pd.Series(r1).reset_index(name="noOfTasks")
r2
```

```python
r23 = r2["lm"].tolist()
r23
```

```python
r21 = r2["SUBJECT"].tolist()
r21
```

```python
r22 = r2["noOfTasks"].tolist()
r22
```

```python
source = ColumnDataSource(data=dict(r21=r21, r22=r22, color=Spectral6))

k = figure(x_range=r21, width=430, height=340, toolbar_location=None,
            tools="hover", tooltips="@r21 @r23: @r22 Tasks", title="noOfTasks by subject by mamatha")

k.vbar(x="r21", top="r22", width=0.6, color="color",legend_field="r21", source=source)

show(k)
```

```python
r["MONTH"] = r["START DATE"].dt.strftime('%m')
rf1= r[(r["MONTH"] == ("10"))]
rf1
```

```python
rf2 = rf1.groupby("SUBJECT")["ACTUAL HOURS"].sum()
rf21 = pd.Series(rf2).reset_index(name="SPEND HOURS")
rf21
```

```python
rf3 = rf21["SUBJECT"].tolist()
rf3
```

```python
rf4 =rf21["SPEND HOURS"].tolist()
rf4
```

```python
source = ColumnDataSource(data=dict(rf3=rf3, rf4=rf4, color=Spectral6))

h = figure(x_range=rf3, width=430, height=500, toolbar_location=None,
            tools="hover", tooltips="@rf3 : @rf4 Hrs", title="spend hours on subject by mamatha in 10th month")

h.vbar(x="rf3", top="rf4", width=0.6, color="color", source=source)

show(h)
```

```python
l = r[(r["START DATE"] >= ("2021-08-13")) & (r["END DATE"] <= ("2021-11-30"))]
l
```

```python
color = ["pink","green","purple","orange"]
l1 = l.groupby(["SUBJECT", "lm"])["ACTUAL HOURS"].sum()
l2 = pd.Series(l1).reset_index(name="Hours")
l2["angle"] = l2["Hours"] / l2["Hours"].sum() *2*pi
l2["color"] = color   # Category20c[len(l2)]
l2
```

```python
o = figure(height=350, toolbar_location=None, tools="hover", tooltips="@NAME,@SUBJECT:@Hours", 
            title="pie chart on hours spend by swapna")


o.wedge(x=1, y=0, radius=0.4, 
        start_angle=cumsum("angle", include_zero=True), end_angle=cumsum("angle"),
        line_color="white", fill_color="color",legend_field="SUBJECT", source=l2)

show(o)
```
