```python
import pandas as pd 
t = pd.read_excel("SwapnaTechWork.xlsx", parse_dates=["START DATE"])
from math import pi
from bokeh.plotting import figure, show, ColumnDataSource
from bokeh.palettes import Spectral6, Category20c
from bokeh.embed import components
from bokeh.transform import cumsum
from bokeh.resources import CDN
```

```python
t
```

```python
t["MONTH"] = t["START DATE"].dt.strftime('%m')
t9= t[(t["MONTH"] == ("10"))]
t9
```

```python
u = t9.groupby(["SUBJECT"])["ACTUALS"].count()
u1 = u.index.tolist()
u1
```

```python
source = ColumnDataSource(data=dict(u1=u1, u=u, color=Spectral6))

v = figure(x_range=u1, width=430, height=340, toolbar_location=None,
            tools="hover", tooltips="@u1: @u", title="noOfTasks of Swapna in 10th month")

v.vbar(x="u1", top="u", width=0.6, color="color",legend_field="u1", source=source)

show(v)
```

```python
script2, div2 = components(v)
cdn_js = CDN.js_files[0]
cdn_js
```

```python
t1 = t.groupby(["SUBJECT", "NAME"])["ACTUALS"].sum()
t2 = pd.Series(t1).reset_index(name="Hours")
t2["angle"] = t2["Hours"] / t2["Hours"].sum() *2*pi
t2["color"] = Category20c[len(t2)]
t2
```

```python
h = figure(height=350, toolbar_location=None, tools="hover", tooltips="@NAME,@SUBJECT:@Hours", 
            title="pie chart on hours spend by swapna")


h.wedge(x=1, y=0, radius=0.4, 
        start_angle=cumsum("angle", include_zero=True), end_angle=cumsum("angle"),
        line_color="white", fill_color="color",legend_field="SUBJECT", source=t2)
```

```python
show(h)
```

```python
s = t[(t["START DATE"] >= ("2021-09-20")) & (t["END DATE"] <= ("2021-10-11"))]
s
```

```python
p=s.groupby(["SUBJECT"])["ACTUALS"].sum()
p1 = p.index.tolist()
p1
```
