```python
import pandas as pd 
g = pd.read_excel("Nareshworksheet.xlsx", parse_dates=["Start Date"])
from math import pi
from bokeh.plotting import figure, show, ColumnDataSource
from bokeh.palettes import Spectral6, Category20c
from bokeh.embed import components
from bokeh.embed import json_item
from bokeh.transform import cumsum
from bokeh.resources import CDN
from bokeh.core.json_encoder import serialize_json
from json.encoder import JSONEncoder
```

```python
g
```

```python
g1 = g.groupby("Subject")["Actual Hours"].sum()
g2 = g1.index.tolist()
g2
```

```python
source = ColumnDataSource(data=dict(g2=g2, g1=g1, color=Spectral6))

y = figure(x_range=g2, width=550, height=630, toolbar_location=None,
            tools="hover", tooltips="@g2: @g1 Hrs", title="Spend Hours on courses of Naresh")

y.vbar(x="g2", top="g1", width=0.6, color="color",legend_field="g2", source=source)

show(y)
```
