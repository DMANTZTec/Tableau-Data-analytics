```python
import pandas as pd
import streamlit as st
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, output_file, show
import matplotlib.pyplot as plt
```

```python
cookiesDf = pd.read_excel('OneDrive-2022-11-30\Cookie Types.xlsx')

ordersDf = pd.read_excel('OneDrive-2022-11-30\Orders.xlsx')

customersDf = pd.read_excel('OneDrive-2022-11-30\Customers.xlsx')
```

```python
namesDf = customersDf[['Customer ID', 'Name']]

# namesDf['Customer ID'] = namesDf['Customer ID'].astype(int)

namesDf
```

```python
df = (ordersDf.groupby('Customer ID').sum()).reset_index()

df.drop(['Order ID'], axis=1, inplace=True)

df
```

```python
customerOrdersDf = pd.merge(namesDf, df, on="Customer ID")

customerOrdersDf
```

```python
ordersCountDf = ordersDf.groupby('Customer ID')['Order ID'].count()

countOfOrders = ordersCountDf.reset_index()

countOfOrders.rename(columns={'Order ID': 'Orders Count'}, inplace=True)

countOfOrders
```

Orders Count $ Profit

```python
customerOrdersDf = pd.merge(customerOrdersDf, countOfOrders, on="Customer ID")

# customerOrdersdf['Revenue'] = customerOrdersdf['Revenue'].astype(int)
# customerOrdersdf['Cost'] = customerOrdersdf['Cost'].astype(int)
# customerOrdersdf['Units Sold'] = customerOrdersdf['Units Sold'].astype(int)

customerOrdersDf['Profit'] = customerOrdersDf['Revenue'] - customerOrdersDf['Cost']

customerOrdersDf
```

Profit Margin

```python
customerOrdersDf['profitMargin'] = customerOrdersDf['Profit'] / customerOrdersDf['Revenue']

customerOrdersDf
```
