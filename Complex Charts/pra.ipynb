```python
exr = data[data["SUBJECT"] == ("KAFKA")]
hours = exr.groupby(["START DATE", "SUBJECT", "TRAINEE"])["SPEND HOURS"].sum()
hoursdata_df = pd.Series(hours).reset_index(name="HOURS")
hoursdata_df
```

```python
hours = hoursdata_df["HOURS"].tolist()
hours
```

```python
S = hoursdata_df[hoursdata_df["TRAINEE"] == ("valisetti suresh")]
suresh = S["TRAINEE"].tolist()
suresh
```

```python
sureshsub= S["SUBJECT"].tolist()
sureshsub
```

```python
dates = S["START DATE"].tolist()
dates
```

```python
sureshhours = S["HOURS"].tolist()
sureshhours
```

```python
N = hoursdata_df[hoursdata_df["TRAINEE"] == ("Naresh Chepuroju")]
naresh = N["TRAINEE"].tolist()
naresh
```

```python
datesi = N["START DATE"].tolist()
datesi
```

```python
subject = hoursdata_df["SUBJECT"].tolist()
subject 
```

```python
source = ColumnDataSource(data=dict(dates = dates, sureshhours = sureshhours,datesi=datesi,nareshhours=nareshhours, 
                                sureshsub=sureshsub, suresh=suresh, color=Spectral7))

spendhours = figure(x_range=dates, width=700, height=350, toolbar_location=None,
            tools="hover")

spendhours.vbar(x="dates", top= "sureshhours", width=0.6, legend_field="dates",color="color", source=source)

spendhours.background_fill_alpha = 0

spendhours.line(x="datesi",y="nareshhours", color="violet",line_width=2,source=source)
 
show(spendhours)
```
